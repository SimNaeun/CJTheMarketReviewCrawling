```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time

# 브라우저 드라이버 설정
driver = webdriver.Chrome()

# 페이지로 이동
url = "https://www.cjthemarket.com/pc/prod/prodDetail?prdCd=40164886&ctgrId=003605&plnId=&giftSetEvntId="
driver.get(url)

# 페이지 로딩 대기
time.sleep(5)  # 페이지 로딩 대기 시간

# "최신순" 정렬 클릭
try:
    latest_sort = WebDriverWait(driver, 10).until(
        EC.element_to_be_clickable((By.XPATH, '//*[@id="prdReview"]/div[4]/div[1]/div[1]/span[2]/a'))
    )
    latest_sort.click()  # 최신순 정렬 클릭
    time.sleep(3)  # 클릭 후 대기
except Exception as e:
    print(f"Failed to click '최신순': {e}")

# 리뷰 수집 함수
review_list = []

def get_reviews():
    try:
        reviews = driver.find_elements(By.XPATH, '//*[@id="reviewTab"]/li/div[2]/div/p/a')
        for review in reviews:
            review_list.append(review.text)
    except Exception as e:
        print(f"Error fetching reviews: {e}")

# 페이지 넘기기 함수
def next_page(page_num):
    try:
        driver.execute_script(f"javascript:goPage({page_num}, 'reviewTab', '40164886', '206899');")
        time.sleep(5)  # 페이지 넘길 때 대기 시간 추가
    except Exception as e:
        print(f"Page {page_num} not found. Error: {e}")

# 첫 50페이지의 리뷰 수집
for page in range(1, 51):
    get_reviews()
    next_page(page + 1)

# 드라이버 종료
driver.quit()

# 수집한 리뷰 출력
for review in review_list:
    print(review)
```
